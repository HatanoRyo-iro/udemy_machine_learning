## 多項式回帰

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
X
```

```python
# データセットを使った線形回帰モデルの訓練
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# データセットを使った多項式回帰モデルの訓練
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
```

```python
X_poly
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
# 結果の可視化（線形回帰）
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='green')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
# 結果の可視化（多項式回帰）
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg2.predict(X_poly), color='green')
```

```python
# 結果の可視化（多項式回帰　より滑らかなカーブ）
X_grid = np.arange(min(X), max(X), 0.1)   # 独立変数を0.1刻み
X_grid = X_grid.reshape(len(X_grid), 1)   # ベクトルを行列に変換

plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color='green')
```

```python
# 線形回帰モデルによる予測
lin_reg.predict([[6.5]])
```
